```python
df=pd.read_csv('book1.csv')
df.head(3)
```

```python
d=df.iloc[:,[1,2,4]]
d1=df.iloc[:,[0,3]]
print(d.head(1))
print(d1.head(1))
```

```python
d=pd.get_dummies(d)
d.head(1)
```

```python
data=pd.concat([d,d1],axis=1)
data.head(1)
```

```python
#data for logistic regression
x=data.iloc[:,[1,2,3,7,8]]
x.head(1)
```

```python
y=data['crime_rec_yes']
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression as glm
model=glm().fit(xtrain,ytrain)
pred_y=model.predict(xtest)
from sklearn.metrics import confusion_matrix as cm
print(cm(ytest,pred_y))
```

```python
data=pd.read_csv('heart.csv')
data.head()
```

```python
x=data.drop(['target'],axis=1)
y=data['target']
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn import tree
m=tree.DecisionTreeClassifier().fit(xtrain,ytrain)
pred_y=m.predict(xtest)
from sklearn.metrics import accuracy_score as au_sc
au_sc(pred_y,ytest)
```

```python
from sklearn.ensemble import RandomForestClassifier as rfc
m=rfc().fit(xtrain,ytrain)
pred_y=m.predict(xtest)
from sklearn.metrics import accuracy_score as au_sc
au_sc(pred_y,ytest)
```

```python
from sklearn.naive_bayes import GaussianNB as nb
m=nb().fit(xtrain,ytrain)
pred_y=m.predict(xtest)
from sklearn.metrics import accuracy_score as au_sc
au_sc(pred_y,ytest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
m=knn.fit(xtrain,ytrain)
pred_y=m.predict(xtest)
from sklearn.metrics import accuracy_score as au_sc
au_sc(pred_y,ytest)
```

```python
from sklearn.svm import SVC
m=SVC(kernel='linear',gamma=1)
m.fit(xtrain,ytrain)
pred_y=m.predict(xtest)
from sklearn.metrics import accuracy_score as au_sc
au_sc(pred_y,ytest)
```
