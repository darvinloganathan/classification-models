```python

import pandas as pd
data=pd.read_csv('book1.csv')
data
x=data['squarefeet']
y=data['price']
import statsmodels.api as sm
model=sm.OLS(y,x).fit()
model.summary()
data['prediction']=model.predict(x)
data
```
